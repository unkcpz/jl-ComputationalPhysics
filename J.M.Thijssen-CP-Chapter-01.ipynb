1.1 In this problem it is assumed that you have at your disposal a routine for solving an ordinary second order differential equation with given initial conditions.

Duffing oscillator:
$$m\ddot{x}(t)=F[x(t),\dot{x}(t),t].$$
$$F[x,\dot{x},t]=-\gamma \dot{x} + 2ax-4bx^3 + F_0 \cos (\omega t).$$

(a) Write a program for the Duffing oscillator and study the motion for different sets of values of the parameters $F_0$, $\omega$, $\gamma$ (use $m=1, a=1/4,b=1/2$) and initial conditions $x_0$ and $\dot{x_0}$.

(b) For the same values of the parameters, print out the values of $x$ and $p=\dot{x}$ each time a period $T=2\pi /\omega$ has elapsed.
Plotting these points shoud yield a resulting curve called the *strange attractor*.

(c) fractal dimentsion

```julia
using DifferentialEquations
using Plots
```

The ODE can be rewrited to a ODE sets:

$$\mathrm{d}x=\dot{x}$$
$$\mathrm{d}\dot{x}=-\gamma \dot{x} + 2ax-4bx^3 + F_0 \cos (\omega t).$$

Let $x(t)$ be `u[1]` and $\dot{x}$ be `u[2]`, then `du[1]` is $\dot{x}(t)$ and `du[2]` is $\ddot{x}(t)$.

(a)

```julia
# m = 1, a = 0, b = 1/4
function duff_osc(du,u,p,t)
    F0, ω, γ = p[1], p[2], p[3]
    du[1] = u[2]
    du[2] = -γ*u[2] + 0 * u[1] - 1 * u[1]^3 + F0 * cos(ω * t)
end
```

```julia
ω = 1.0
u0 = [0.5, 0.0]
tspan = (0, 150.0)
p = [7.5, ω, 0.05]
prob = ODEProblem(duff_osc, u0, tspan, p)
sol = solve(prob,Tsit5());
```

```julia
plot(sol, vars=(0,1), linewidth=5,title="Solution to duffing oscilation", 
    xaxis="Time (t)", yaxis="u(t)", label="x0 = 0.5000");
```

```julia
ω = 1.0
u0 = [0.5001, 0.0]
tspan = (0, 150.0)
p = [7.5, ω, 0.05]
prob = ODEProblem(duff_osc, u0, tspan, p)
sol = solve(prob,Tsit5());
```

```julia
plot!(sol, vars=(0,1), lw=3,ls=:dash,label="x0=0.5001")
```

```julia
function duff_osc(du,u,p,t)
    F0, ω, γ = p[1], p[2], p[3]
    du[1] = u[2]
    du[2] = -γ*u[2] - 1 * u[1] - 5 * u[1]^3 + F0 * cos(ω * t)
end
```

```julia
ω = 0.5
u0 = [1.5, 0.0]
tspan = (0, 100000.0)
p = [8.0, ω, 0.02]
prob = ODEProblem(duff_osc, u0, tspan, p)
sol = solve(prob, saveat=2π/ω);
```

```julia
plot(sol, vars=(1,2), seriestype=:scatter, markersize=[0.5,0.5],title="Strange attractor for the Duffing oscillator", 
    xaxis="x", yaxis="p", label="", xlims=[1,2])
```

1.2 In this problem, we consider diffusion limited aggregation.

(a) Write a program for generating DLA clusters on a square lattice of size 150 $\times$ 150. Generate a cluster of about 9000 sites, and write the sites occupied by this cluster ot a file for viewing using a graphics program.

(b) fractal dimension

(a)

```julia
n = 150
n_points = 9000
dsites = fill!(Vector{Tuple{Int64, Int64}}(undef, n_points), (-1, -1));
```

```julia
dsites[1] = (div(n, 2), div(n, 2));
```

```julia
function rand_boundary(n)
    b = rand(1:4)
    bs = rand(1:n)
    if b == 1
        site_b = (0, bs)
    elseif b == 2
        site_b = (n, bs)
    elseif b == 3
        site_b = (bs, 0)
    else
        site_b = (bs, n)
    end
    site_b
end
```

```julia
function move(site)
    sites = [site .+ m for m in [(0, 1), (1, 0), (0, -1), (-1, 0)]]
    sites = [s for s in sites if all((0, 0) .<= s .<= (n, n))]
    new_site = rand(sites)
end
```

```julia
for i = 2:n_points
    save = now_site = rand_boundary()
    while now_site ∉ dsites[1:i-1]
        save = now_site
        now_site = move(now_site)
    end
    dsites[i] = save
end
```

```julia
I = [d[1] for d in dsites]
J = [d[2] for d in dsites];
```

```julia
scatter(I[1:20], J[1:20], markersize=1, shape=:auto, xlims=(0, 150), ylims=(0,150), size=(500, 500))
```

```julia
anim = @animate for i = 1:9000
    scatter(I[1:i], J[1:i], markersize=1, shape=:rect, xlims=(0, 150), ylims=(0,150), size=(500, 500), label="")
end
gif(anim, "./DAL_fps30.gif", fps = 30)
```
