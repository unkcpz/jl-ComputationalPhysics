```julia
using DifferentialEquations
using Plots
using LaTeXStrings
```

```julia
function sch(du, u, p, r)
    du[1] = u[2]
    du[2] = (r^2 - 3) * u[1]
end
```

```julia
rspan = (0.0, 2.0)
u0 = [0.0, 1.0]
prob = ODEProblem(sch, u0, rspan)
```

```julia
h = 0.0001
sol = solve(prob, saveat=h);
```

```julia
plot(sol, vars=(0,1), xaxis="r", yaxis="u(r)", label="u(r)", w=10);
```

```julia
h = 0.0001
A = exp(h^2/2)
f(r) = A * r * exp(-r^2/2)
```
