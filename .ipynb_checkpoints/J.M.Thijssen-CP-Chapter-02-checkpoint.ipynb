```julia
using Plots
```

```julia
l, E = 0, 3.0
h = 0.001
V(r) = r^2
```

```julia
g(r, l, E) = V(r) + l*(l+1) / r^2 - E
```

```julia
function radi(l, h, rmax, E)
    rlast = rmax + 2π/√E
    m = Int(floor(rlast / h))
    mmax = Int(floor(rmax / h))
    F = zeros(m)
    R = zeros(m)
    G = zeros(m)
    F[1] = 0.000
    F[2] = 0.001
    R[1] = eps()
    R[2] = h
    G[1] = g(R[1], l, E)
    G[2] = g(R[2], l, E)
    hh = h*h / 12
    for i = 2: m-1
        R[i+1] = R[i] + h
        G[i+1] = g(R[i+1], l, E)
        F[i+1] = (-F[i-1] + 2*F[i] + 10*G[i]*F[i]*hh + G[i-1]*F[i-1]*hh) / (1 - G[i+1]*hh)
    end
 
    (R=R, F=F, r1=rmax, r2=rlast, ur1=F[mmax], ur2=F[m])
end
```

```julia
res = radi(l, 0.001, 1.0, E)
plot(res.R, res.F, label="numerov method", w=10);
```

```julia
@show res.r1, res.r2, res.ur1, res.ur2
```
