```julia
using DifferentialEquations
using Plots
using Roots
```

```julia
function radi(l, h, rmax, E)
    m = Int(ceil(rmax / h))
    F = zeros(m)
    R = zeros(m)
    G = zeros(m)
    F[1] = 0.000
    F[2] = 0.001
    R[1] = eps()
    R[2] = R[1] + h
    G[1] = l*(l+1)/(R[1]^2)-2/R[1]-2*E
    G[2] = l*(l+1)/(R[2]^2)-2/R[2]-2*E
    hh = h*h / 12
    for i = 2: m-1
        R[i+1] = R[i] + h
        G[i+1] = l*(l+1)/(R[i+1]^2)-2/R[i+1]-2*E
        F[i+1] = (-F[i-1] + 2*F[i] + 10*G[i]*F[i]*hh + G[i-1]*F[i-1]*hh) / (1 - G[i+1]*hh)
    end
    
    Radials = F ./ R
    
    rsave = rmax - 2Ï€/4
    msave = Int(ceil(rsave/h))
    integral = sum(Radials[msave:m]) * h
    
    (R=R, F=F, Radials=Radials, integral=integral)
end
```

```julia
x = -0.125
R, F, Radials = radi(0, 0.001, 10+abs(1/x), x)
plot(R, F, w=10)
plot!(R, Radials, w=3)
# plot!(x->10x*exp(-x), 0, 12)
```

```julia
function eintfar(x)
    res = radi(0, 0.001, 10+abs(1/x), x)
    return res.integral
end
```

### ODE Problem

```julia
function sch(du, u, p, t)
    E = p[1]
    du[1] = u[2]
    du[2] = (-2E) * u[1]
end
```

```julia
rspan = (0.01, 1.0)
p = [4]
u0 = [0.0, 1.0]
prob = ODEProblem(sch, u0, rspan, p)
```

```julia
sol = solve(prob, Rodas4())
plot(sol)
```

```julia
sol.u[end]
```

```julia
function poisson(du, u, p, r)
    du[1] = u[2]
    du[2] = -4 * r * exp(-2r)
end
```

```julia
h = 0.001
u0 = [0.0, h]
rspan = (0, 10.0)
prob = ODEProblem(poisson, u0, rspan)
```

```julia
sol = solve(prob);
```

```julia
u = [i[1] for i in sol.u]
Î± = (1 - u[end]) / sol.t[end]
u .= u .+ sol.t * Î±;
```

```julia
plot(sol.t, u, w=10);

hatree_(r) = -(r+1) * exp(-2*r) + 1
plot!(hatree_, 0, 10, line=(:dash, 3))
```
